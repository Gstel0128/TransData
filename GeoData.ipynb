```python
df = pd.read_table(r'/Users/Kevin/Desktop/FinalProject/SiouxFalls/SiouxFalls_trips.txt',encoding = 'utf8',sep ='\t',header = None)
df = df[0].str.split(';',expand = True)
df.head(20)
```

```python
d1 = {}
tmp = {}
for i in range(0,144):  
    if (i%6 != 0):
        for j in range(0,5):
            if (df.iloc[i,j] != ' ') :
                key = int(df.iloc[i,j].split(':')[0].strip())
                value = float(df.iloc[i,j].split(':')[1].strip())
                tmp[key] = value
    if  (i+1)%6 == 0:
        d1[i// 6] = tmp
        tmp = {}

```
